```python
df=pd.read_csv('car_fuel_efficiency.csv')
df.head()
```

```python
#Q2. Records count
df.info()
#ANS: 9704
```

```python
#Q3. Fuel types
df['fuel_type'].unique()
#ANS 2 fuel types
```

```python
#Q4. Missing values
df.isnull().sum()
#ANS: 4 columns/input features with missing values
```

```python
#Q5. Max fuel efficiency
df[df['origin']=='Asia']['fuel_efficiency_mpg'].max()
#ANS: 23.75
```

```python
#Q6. Median value of horsepower
median=df['horsepower'].median()
print(median)#149hp
mode=df['horsepower'].mode()[0]
print(mode)#152hp
df['horsepower'].fillna(mode,inplace=True)
new_median=df['horsepower'].median()
print(new_median)#152hp
#ANS: yes the median has changed and it has increased
```

```python
#Q7. Sum of weights

newdf=df.loc[df['origin'] == "Asia", ['vehicle_weight', 'model_year']]
X=newdf.head(7)
XTX=X.T@X
XTX_inv = np.linalg.inv(XTX) 
y=[1100, 1300, 800, 900, 1000, 1100, 1200]
w=XTX_inv@X.T@y
print(w.sum())
#ANS:0.51
```
